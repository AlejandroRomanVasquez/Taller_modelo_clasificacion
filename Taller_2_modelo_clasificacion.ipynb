# Preprocesamiento

Selección de variables para la creación del modelo

```python
df_titanic_red = df_titanic[['Survived', 'Pclass',
                             'Sex','Age','Fare',]]
print(df_titanic_red.shape)
```

Eliminación de filas con datos faltantes

```python
# Eliminar filas con datos faltantes
df_titanic_red_completo = df_titanic_red.dropna()
df_titanic_red_completo.info()
```

Carga de paquetes para graficar

```python
# Gráfica de sobrevivientes por Pclass
sns.countplot(x='Survived', hue='Pclass',
              data=df_titanic_red_completo)
plt.title("Supervivencia por Clase")
plt.show() #Mostrar el gráfico
```

Generación de variales ficticias para las variables categóticas 'Sex' y 'Pclass'

```python
# Convertir 'Sex' y 'Pclass' a variables ficticias
df_titanic_ficticias = pd.get_dummies(df_titanic_red_completo,
                                                   columns=['Sex','Pclass'], drop_first=True)
df_titanic_ficticias
```

```python
#Se transforma cada columna
df_titanic_ficticias[['Sex_male']] = df_titanic_ficticias[['Sex_male']].astype(int)
df_titanic_ficticias[['Pclass_2']] = df_titanic_ficticias[['Pclass_2']].astype(int)
df_titanic_ficticias[['Pclass_3']] = df_titanic_ficticias[['Pclass_3']].astype(int)

df_titanic_ficticias
```

# Generación de conjunto de entrenamiento y prueba

Separar los datos en entrenamiento y prueba

```python
# Separar variables predictoras (X) y variable respuesta (y)
X = df_titanic_ficticias.drop(columns=['Survived'])
y = df_titanic_ficticias['Survived']
```

Marco de dato con puras variables predictoras

```python
X
```

Marco de datos con solo la respuesta

```python
y
```

```python
#Necesito importar la función train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_entr, X_prue, y_entr, y_prue = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
y_entr.value_counts(normalize=True)
```

```python
y_prue.value_counts(normalize=True)
```

# Ajuste de modelo de regresión logística

```python
#Importación de clase LogisticRegression
from sklearn.linear_model import LogisticRegression

#Creación de objeto tipo LogisticRegression
modelo_logistico = LogisticRegression()
```

```python
#Ajuste del modelo
modelo_logistico.fit(X_entr, y_entr)
```

```python
df_coef = pd.DataFrame({
    'Variable': X.columns,
    'Coeficiente': modelo_logistico.coef_.flatten()
})
df_coef
```

# Evaluación del modelo

```python
X_prue.head(5)
```

Visualizar predicciones

```python
# Predicciones
y_pred = modelo_logistico.predict(X_prue)
print(y_pred)
```

```python
#Creación de marco de datos con predicciones
df_predicciones = pd.DataFrame({
    'Real': y_prue,
    'Prediccion': y_pred
})
df_predicciones
```

```python
res_predicciones = pd.concat([X_prue, df_predicciones], axis=1)
res_predicciones
```

```python
#Importación de funciones para evaluar el modelo
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Evaluar el modelo
print("Matriz de confusión:")
print(confusion_matrix(y_prue, y_pred))
```

```python
print("Exactitud del modelo:")
round(accuracy_score(y_prue, y_pred), 4)
```
